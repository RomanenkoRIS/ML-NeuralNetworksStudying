```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
import seaborn as sns
import torch
import torch.functional as F

%matplotlib inline
from pylab import rcParams
rcParams['figure.figsize'] = 12, 6
```

\begin{equation}
     {\sigma(x) = \frac{1}{{1+e^{-x}}}}
\end{equation}

![image.png](1.png)

```python
def sigmoid(x):
    return 1 / (1 + torch.exp(-x))


def cost(X, y, theta):
    
    N = len(y)
    H = sigmoid(X @ theta)
    cost = 1/N * (((-y).T @ torch.log(H)) - ((1-y).T @ torch.log(1-H)))
    return cost[0]


def gradient_descent(X, y, params, learning_rate, iterations):
    m = y.size()[0]
    cost_history = torch.zeros((iterations,1),requires_grad=False)
    
    for i in range(iterations):
        params = params - (learning_rate/m) * (X.T @ (sigmoid(X @ params) - y))
        cost_history[i] = cost(X, y, params)[0]
    
    return (cost_history, params)


def predict(X, params):
    return torch.round(sigmoid(X @ params))
```

```python
X, y = make_classification(n_samples=1000, n_features=2, n_redundant=0, n_informative=1,
                             n_clusters_per_class=1, random_state=14)

y = y[:,np.newaxis]
X = torch.tensor(X, requires_grad=False,dtype=torch.double)
y = torch.tensor(y, requires_grad=False,dtype=torch.double)
```

```python
m = y.size()[0]
X = torch.hstack((torch.ones((m,1)),X),)
n = X.size(1)
params = torch.zeros((n,1),requires_grad=False, dtype=torch.double)

iterations = 3500
learning_rate = 0.05

initial_cost = cost(X, y, params)

print("Initial Cost is: {} \n".format(initial_cost[0]))

(cost_history, params_optimal) = gradient_descent(X, y, params, learning_rate, iterations)
print("Optimal Parameters are: \n", params_optimal, "\n")

plt.figure()
sns.set_style('white')
plt.plot(range(len(cost_history)), cost_history, 'r')
plt.title("Convergence Graph of Cost Function")
plt.xlabel("Number of Iterations")
plt.ylabel("Cost")
plt.show()
```

```python
y_pred = predict(X, params_optimal)
score = float(sum(y_pred == y))/ float(len(y))

print(score)

slope = -(params_optimal[1] / params_optimal[2])
intercept = -(params_optimal[0] / params_optimal[2])

sns.set_style('white')
sns.scatterplot(X[:,1],X[:,2],hue=y.reshape(-1));

ax = plt.gca()
ax.autoscale(False)
x_vals = np.array(ax.get_xlim())
y_vals = intercept + (slope * x_vals)
plt.plot(x_vals, y_vals, c="k")
```

```python
def gen_params(grad, n):
    w = 0.2 * torch.rand(n, 1, requires_grad = grad,dtype=torch.float) - 0.1
    b = 0.2 * torch.rand(1, requires_grad = grad,dtype=torch.float) - 0.1
    return w, b


def f(x, w_):
    return torch.sigmoid(torch.matmul(x, w_))
```

```python
n = 10
m = 1000
x = torch.rand(m, n,requires_grad =True, dtype=torch.float)

x_train = x[:int(m*0.8)]
x_test = x[int(m*0.8):]

orig_w, orig_b = gen_params(True, n)

y = f(x, orig_w)

y_train = (y[:int(m*0.8)]>0.5).float()
y_test = (y[int(m*0.8):]>0.5).float()

plt.plot(x_train.tolist(),y_train.tolist(),'o',color='green',markersize=5)
plt.show()
```

```python
initial_cost = cost(x_train, y_train, orig_w)

print("Initial Cost is: {} \n".format(initial_cost[0]))

(cost_history, params_optimal) = gradient_descent(x_train, y_train, orig_w, 0.03, 10000)
# print("Optimal Parameters are: \n", params_optimal, "\n")

plt.figure()
sns.set_style('white')
plt.plot(cost_history.tolist(), 'r')
plt.title("Convergence Graph of Cost Function")
plt.xlabel("Number of Iterations")
plt.ylabel("Cost")
plt.show()
```

```python
# Train set
y_pred_tr = predict(x_train, params_optimal)
score_tr = sum(y_pred_tr == y_train)/ (y_train).size()[0]

# Test set
y_pred_te = predict(x_test, params_optimal)
score_test = sum(y_pred_te == y_test)/ (y_test).size()[0]

print(f'Train set: {round(float(score_tr[0]),3)}\nTest set: {round(float(score_test[0]),3)}')

params_optimal
```
